polynomail regression

```
import numpy as np
import matplotlib.pyplot as plt
x=np.array([0,1,2,3,4,5,6])
y=np.array([0.2,1.1,1.3,1.2,2.1,1.9,2.5])
#plt.plot(x,y,'*')

coefi1=np.polyfit(x,y,1)
coefi2=np.polyfit(x,y,2)
coefi3=np.polyfit(x,y,3)
coefi4=np.polyfit(x,y,4)
coefi5=np.polyfit(x,y,5)
coefi6=np.polyfit(x,y,6)

print(coefi1)
print(coefi2)
print(coefi3)
print(coefi4)
print(coefi5)
print(coefi6)

plt.plot(x,y,'*')

plt.plot(x,np.polyval(coefi1,x),'black')
plt.plot(x,np.polyval(coefi2,x),'g')
plt.plot(x,np.polyval(coefi3,x),'r')
plt.plot(x,np.polyval(coefi4,x),'y')
plt.plot(x,np.polyval(coefi5,x),'c')
plt.plot(x,np.polyval(coefi6,x),'blue')
```
